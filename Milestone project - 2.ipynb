```python
import random

suit = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
rank = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
value = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}

playing = True
```

```python
class Card():
    
    def __init__(self,suit,rank):
        self.suit = suit
        self.rank = rank
        
    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class Deck:
    
    def __init__(self):
        
        self.the_deck = []
        
        for suits in suit:
            for ranks in rank:
                
                self.the_deck.append(Card(suits,ranks))
                
    def shuffle(self):
        
        random.shuffle(self.the_deck)
        
    def deal(self):
        
        return self.the_deck.pop()
    
    def __str__(self):
        deck_comp = ""
        for card in self.the_deck:
            deck_comp += "\n" + card.__str__()
            
        return "the deck has "+ deck_comp
        
```

```python
class Hand:
    def __init__(self):
        self.cards = []  # start with an empty list as we did in the Deck class
        self.value = 0   # start with zero value
        self.aces = 0    # add an attribute to keep track of aces
    
    def add_card(self,card):
        
        self.cards.append(card)
        self.value += value[card.rank]
        
        if rank == "Ace":
            self.aces += 1
    
    def adjust_for_ace(self):
            while self.value > 21 and self.aces > 0:
                self.value -=10
                self.aces -= 1
```

```python
class Chips:
    
    def __init__(self):
        self.total = 100  # This can be set to a default value or supplied by a user input
        self.bet = 0
        
    def __str__(self):
        print(f'you currently have {self.total}')
        
    def win_bet(self):
        
        self.total += self.bet 
    
    def lose_bet(self):
        self.total -= self.total
        
    
```

```python
def take_bet(chips):
    
    while True:
        try:
            chips.bet = int(input("Please keep a bet: "))
        
        except:
            
            print("not a number!")
            
            continue
            
        else:
            if chips.bet > chips.total:
                print("not sufficient funds")
                continue 
            else:
                print("bet placed")
                break
            
            
```

```python
def hit(deck,hand):
    hand.add_card(deck.deal())
    hand.adjust_for_ace()       
```

```python
def hit_or_stand(deck,hand):
    global playing  # to control an upcoming while loop
    
    playornot = True
    
    while playornot:
        playornot = input("Do you want a hit? Y or N ").upper
        
        if playornot == "Y":
            hit(deck,hand)
            
        elif playornot == "N":
            print("dealers turn! player stands")
            playing = False
            
        else:
            print("Error!")
            playing = True
            
        break
```

```python
def show_some(player,dealer):
    print("The players cards")
    for cards in [player.cards[0],player.cards[1]]:
        print(cards)
    
    print("Dealers cards one hidden")
        
    print(dealer.cards[1])
    
def show_all(player,dealer):
    
    print("Players hand: ")
    
    for cards in player.cards:
        print(cards)
    print(player.value)
    
    print("Dealers hand: ")
    
    for cards in dealer.cards:
        print(cards)
    print(dealer.value)
```

```python
def player_busts(player,dealer,chips):
    print("Player lost the bet!")
    chips.lose_bet()

def player_wins(player,dealer,chips):
    print("Player won!")
    chips.win_bet()

def dealer_busts(player,dealer,chips):
    print("Dealer bust! Player won the bet!")
    chips.win_bet()
    
def dealer_wins(player,dealer,chips):
    print("Dealer won!")
    chips.lose_bet()
    
def push():
    print("No one won ")
```

```python
while True:
    # Print an opening statement
    print("Welcome to play the blackjack game!")

    
    # Create & shuffle the deck, deal two cards to each player
    thedeck = Deck()
    thedeck.shuffle()  
    
    playerhand = Hand()
    playerhand.add_card(thedeck.deal())
    playerhand.add_card(thedeck.deal())
    
    dealerhand = Hand()
    dealerhand.add_card(thedeck.deal())
    dealerhand.add_card(thedeck.deal())
    
    # Set up the Player's chips
    
    playerchips = Chips()
    
    
    # Prompt the Player for their bet
    
    take_bet(playerchips)

    
    # Show cards (but keep one dealer card hidden)
    show_some(playerhand,dealerhand)

    
    while playing:  # recall this variable from our hit_or_stand function
        
        # Prompt for Player to Hit or Stand
        hit_or_stand(thedeck,playerhand)
        
        # Show cards (but keep one dealer card hidden)
        
        show_some(playerhand,dealerhand)
 
        
        # If player's hand exceeds 21, run player_busts() and break out of loop
        if playerhand.value > 21:
            player_busts(playerhand,dealerhand,playerchips)
            break

    # If Player hasn't busted, play Dealer's hand until Dealer reaches 17
        if playerhand.value < 21 and playerhand.value != 21:
            while dealerhand.value < 17:
                hit(thedeck,dealerhand)
                if dealerhand.value == 17:
                    break
                else:
                    continue
                
        # Show all cards
        
        show_all(playerhand,dealerhand)
    
        # Run different winning scenarios
        if playerhand.value == 21 or playerhand.value > dealerhand.value:
            player_wins(playerhand,dealerhand,playerchips)
            break
            
        elif dealerhand.value > 21:
            dealer_busts(playerhand,dealerhand,playerchips)
            break
            
        elif dealerhand.value == 21 or dealerhand.value > playerhand.value:
            dealer_wins(playerhand,dealerhand,playerchips)
            break
        else:
            push()
            break
        
    
    # Inform Player of their chips total 
        print(playerchips.total)
    
    
    # Ask to play again
    answer = input("Do you want to play again? Y or N ").upper()
    
    if answer == "Y":
        continue
    elif answer == "N":
        print("Thank you for playing!")
        break
```
